# `Numpy`

`Numpy` 使用 `ndarray` 对象来处理多维数组, 该对象是一个快速而灵活的大数据容器.

n: 任意个, d: dimension, array: 数组

`ndarray` 支持向量化运算, `Numpy` 底层使用 `C` 语言编写, 内部解除了 GIL (全局解释器锁), 其对数组的操作速度不受 `Python` 解释器的限制, 效率远高于纯 `Python` 代码.


```python
score = np.array([
    [80, 89, 86, 67, 79],
    [78, 97, 89, 67, 81],
    [90, 94, 78, 67, 74],
    [91, 91, 90, 67, 69],
    [76, 87, 75, 67, 86],
    [70, 79, 84, 67, 84],
    [94, 92, 93, 67, 64],
    [86, 85, 83, 67, 80]
    ])

score  # 嵌套生成二维数组
```

## 常用属性

| 属性               | 解释                       |
| ------------------ | -------------------------- |
| `ndarray.shape`    | 数组维度的元组             |
| `ndarray.ndim`     | 数组维数                   |
| `ndarray.size`     | 数组中元素数量             |
| `ndarray.itemsize` | 一个数组元素的长度（字节） |
| `ndarray.dtype`    | 数组元素的类型             |



```python
score.shape  # (8, 5)
score.ndim  # 2
score.size  # 40
score.dtype  # dtype('int32')

a = np.array([1.1, 1.2, 2.2])
a.dtype  # float64
```

## 基本操作

### 生成数组

```python
# 生成数组

# 生成 0 1 数组
np.zeros(shape=(3, 4), dtype='float32')
np.ones(shape=[2, 3], dtype=np.int32)

# 从现有数组生成
data1 = np.array(score)  # 深拷贝
data2 = np.asarray(score)  # 浅拷贝
data3 = np.copy(score)  # 深拷贝

# 生成指定范围的数组
np.linspace(0, 10, 100)  # 闭区间, 等距离
np.arange(0, 10, 5)  # 5 为步长
```

```python
# 生成随机数组
data4 = np.random.uniform(low=-1, high=1, size=1000000)
# 生成均匀分布并检验
plt.figure(figsize=(20, 8), dpi=600)
plt.hist(x=data4, bins=1000)
plt.show()
```

```python
# 正态分布
data5 = np.random.normal(loc=1.75, scale=0.1, size=1000000)
plt.figure(figsize=(20, 8), dpi=600)
plt.hist(x=data5, bins=1000)
plt.show()
```

```python
# 随机模拟 8 只股票2周的涨幅数据
stock_change = np.random.normal(loc=0, scale=1, size=(8, 10))
stock_change[stock_change > 0.5] = 1.1
stock_change
```

### 数组索引

```python
stock_change[0, :3]
```

### 形状修改

```python
a = stock_change.reshape([10, 8])  # 返回新的 ndarray
stock_change.resize((10, 8))  # 原地修改
stock_change.T;  # 转置
```

### 类型修改

```python
stock_change.astype("int32")
stock_change.tostring() # ndarray序列化到本地
```

### 数组去重

```python
temp = np.array([[1, 2, 3, 4], [3, 4, 5, 6]])
np.unique(temp)
set(temp.flatten())
```

## `ndarray` 运算

### 逻辑运算

```python
stock_change[stock_change > 0.5] = 2
np.all(stock_change[0:2, 0:5] > 0)
np.any(stock_change[:5, :] > 0)
```

`numpy.hstack` 水平拼接

`numpy.vstack` 竖拼接

### 矩阵运算

`np.mat()` 数组转换成矩阵类型


```python
data1 = np.array([[1, 2, 3], [3, 4, 5]])
data1 = np.mat(data1)
data2 = np.array([[1, 3], [4, 5], [2, 3]])
data2 = np.mat(data2)
data1 * data2
```
