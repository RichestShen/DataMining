## 折线图

### 设置画布属性

 `plt.figure()`

`dpi`: 图像清晰度

`figsize`: 指定图像的长宽, 用元组类型表示.

### 保存图像

`plt.savefig()`

必须写在 `plt.show()` 之前.

### 完善折线图

修改 `x, y` 刻度

`plt.xticks(x, **kwargs)`

`plt.yticks(y, **kwargs)`

添加网格显示

`plt.grid()`

添加描述信息

`plt.title()`

`plt.xlabel()`, `plt.ylabel()`

```python
# 准备数据
import random
x = range(60)
temp_shanghai = [random.uniform(15, 18) for i in x]
temp_beijing = [random.uniform(1, 3) for i in x]

# 创建画布
plt.figure(figsize=(20, 8), dpi=300)

# 绘制图像
plt.plot(x, temp_shanghai, color='r', label='上海')
plt.plot(x, temp_beijing, color='b', label='北京')

# 修改刻度
x_label = [f'11点{i}分' for i in x]
plt.xticks(x[::5], x_label[::5])  # 注意步长一一对应
# plt.yticks(range(40)[::5])  # 步长
plt.yticks(range(0, 40, 5))

# 添加网格
plt.grid(True, linestyle='--', alpha=0.5)

# 添加描述信息
plt.xlabel('时间')
plt.ylabel('温度')
plt.title('上海北京温度变化')
plt.legend()  # 图例显示

# 显示图像
plt.show()
```

### 多坐标系图(OOP 方法)



```python
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(30, 8), dpi=300)

# 绘制图像
axes[0].plot(x, temp_shanghai, color='r', label='上海')
axes[1].plot(x, temp_beijing, color='b', label='北京')

# 修改刻度
x_label = [f'11点{i}分' for i in x]

axes[0].set_xticks(x[::5], x_label[::5])  # 注意步长一一对应
# plt.yticks(range(40)[::5])  # 步长
axes[0].set_yticks(range(0, 40, 5))
axes[1].set_xticks(x[::5], x_label[::5])  # 注意步长一一对应
axes[1].set_yticks(range(0, 40, 5))

axes[0].legend()  # 图例显示
axes[1].legend()

# 添加网格
axes[0].grid(True, linestyle='--', alpha=0.5)
axes[1].grid(True, linestyle='--', alpha=0.5)

# 添加描述信息
axes[0].set_xlabel('时间')
axes[0].set_ylabel('温度')
axes[0].set_title('上海温度变化')

axes[1].set_xlabel('时间')
axes[1].set_ylabel('温度')
axes[1].set_title('北京温度变化')

# 显示图像
plt.show()
```
